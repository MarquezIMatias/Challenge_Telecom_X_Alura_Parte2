Método Chi-Squared para revisión de variables categóricas

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import chi2

# Crea copia
df_rank = df.copy()

# Convertir Churn para binario
df_rank["Churn"] = df_rank["Churn"].map({"Yes": 1, "No": 0})

# Separar variables categóricas
categorical_cols = df_rank.select_dtypes(include=["object"]).columns

# Aplicar Label Encoding temporario
le = LabelEncoder()
for col in categorical_cols:
    df_rank[col] = le.fit_transform(df_rank[col])

# Separar X e y
X_cat = df_rank[categorical_cols]
y = df_rank["Churn"]

# Aplicar teste Qui-cuadrado
chi_scores, p_values = chi2(X_cat, y)

# Criar DataFrame com ranking
chi_df = pd.DataFrame({
    "Variable": categorical_cols,
    "Chi2 Score": chi_scores,
    "p-value": p_values
})

# Ordenar por importancia
chi_df = chi_df.sort_values(by="Chi2 Score", ascending=False)

chi_df
```

```python
# Se deben eliminar todas las variables categóricas cuyo p-value sea menor a 0.05

significant_vars = chi_df[chi_df["p-value"] < 0.05]

print("\n✅ Variables significativas (p < 0.05):")
display(significant_vars)
```

```python
# Con el análisis anterior, solamente 2 variables categóricas serán eliminadas
# Estas variables son: "Gender" y "PhoneService"

df = df.drop(columns=["Gender", "PhoneService"])
df
```

### Modelo 1: Regresión Logísitca

```python
# importamos la biblioteca para el modelo, y también las bibliotecas para los reportes, matriz de confusión y puntaje de precisión.

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Instanciar el modelo

log_model = LogisticRegression(max_iter=1000)

# Entrenar el modelo con los datos NORMALIZADOS

log_model.fit(X_train_norm, y_train)

# Realizar las predicciones con los datos NORMALIZADOS

y_pred_log_norm = log_model.predict(X_test_norm)

# Evaluación

print("--- REGRESIÓN LOGÍSTICA (Datos Normalizados) ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_log_norm):.4f}")
print("\nMatriz de Confusión:")
print(confusion_matrix(y_test, y_pred_log_norm))
print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred_log_norm))
```

Modelo de Regresión Logísitca Balanceado

```python
# Instanciamos con balanceo de clases simple (class_weight)

log_model_balanced = LogisticRegression(max_iter=1000, class_weight='balanced')

# Entrenamos con los mismos datos NORMALIZADOS

log_model_balanced.fit(X_train_norm, y_train)

# Predicciones con los datos NORMALIZADOS

y_pred_log_bal = log_model_balanced.predict(X_test_norm)

# Evaluación

print("--- REGRESIÓN LOGÍSTICA BALANCEADA ---")
print(f"Accuracy: {accuracy_score(y_test, y_pred_log_bal):.4f}")
print("\nMatriz de Confusión:")
print(confusion_matrix(y_test, y_pred_log_bal))
print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred_log_bal))
```

#### 2. Principales Factores de Cancelación

Basándonos en la importancia de variables del Random Forest y los coeficientes de la Regresión Logística, identificamos tres pilares que disparan el Churn:

1. **Fragilidad en Clientes Nuevos (Tenure):** La antigüedad es el predictor más fuerte. Existe una "zona de alto riesgo" en los primeros 12 meses donde la probabilidad de fuga es máxima.
2. **El "Efecto Fibra Óptica":** Los clientes con servicio de Fibra Óptica tienen un riesgo significativamente mayor de cancelar en comparación con los de DSL. Esto sugiere una posible desconexión entre el precio, la expectativa y la realidad del servicio entregado.
3. **Métodos de Pago y Contratos:** Los contratos con cargo mensual y el uso de Cheque Electrónico son facilitadores de la fuga. Los clientes con estos perfiles no tienen barreras de salida ni incentivos de permanencia, por lo que se les hace muy fácil cancelar.


#### 3. Estrategias de Retención Propuestas

Se sugieren las siguientes acciones estratégicas:

* **Programa de acompañamiento para nuevos clientes "Kit de Bienvenida":** Implementar seguimiento frecuente durante los primeros 6 meses de contrato, ofreciendo soporte técnico proactivo, encuestas de satisfacción e información sobre el servicio contratado.
* **Revisión del Servicio de Fibra Óptica:** Investigar la causa técnica o comercial del Churn por este servicio. Si es por precio, crear planes de fidelización; si es técnico, reforzar el soporte y evaluar el estado de la tecnología vs la competencia.
* **Incentivos para pasarse a un modelo anual o bianual:** Ofrecer descuentos y/o beneficios exclusivos (como streaming gratis, por ejemplo) a clientes que solo tiene el servicio de modalidad mensual.
* **Promoción de Pago Automático (PAD o PAC):** Incentivar el cambio de Cheque Electrónico a Tarjeta de Crédito o Transferencia Automática, ofreciendo un detalle de bajo costo como un día de streaming gratis o un pequeño descuento en la siguiente boleta, ya que estos métodos suelen generar una relación más estable y duradera.


#### 4. Conclusión

La diferencia principal entre modelos (Tenure en RF vs. Fibra en Logística) nos confirma que el Churn es un fenómeno que se atribuye a múltiples causas, por lo que no puede ser observado como un fenómeno aislado: mientras que el tiempo de permanencia define la estructura de la lealtad, el tipo de servicio (Fibra) actúa como el detonante principal de la insatisfacción actual.
